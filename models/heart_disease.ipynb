```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
heart_df = pd.read_csv('../dataset/heart.csv')
```

```python
heart_df.head()
```

```python
heart_df.describe()
```

```python
# separating features and targets
X = heart_df.drop(columns='target',axis=1)
Y = heart_df.target
```

```python
# Test and train split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,stratify=Y,test_size=0.2,random_state=2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
data = np.array([63,	1,	3,	145,	233,	1,	0,	150,	0,	2.3,	0,	0,	1])
model.predict(data.reshape(1,-1))
```

```python
# saving the model
import pickle
```

```python
filename = '../trained_model/heart_disease_trained_model.sav'
pickle.dump(model, open(filename, 'wb'))
```
